```python
#create a new column 'age_group' based on age
def age_group(age):
    if age < 18:
        return 'child'
    elif age < 60:
        return 'adult'
    else:
        return 'senior'

df['age_group'] = df['age'].apply(age_group)
df.head()
```

```python
#create a new column 'fare_group' based on fare
def fare_group(fare):
    if fare < 20:
        return 'low'
    elif fare < 50:
        return 'medium'
    else:
        return 'high'
df['fare_group'] = df['fare'].apply(fare_group)
df.head()
```
